```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```python
# numpy를 이용하여 구현하세요.
# AND 게이트를 구현하시오. 입력 2개 가중치 2개(w1= 0.5, w2= 0.5), bias = -0.7로 선형 결합 결과가 0보다 작으면 0, 아니면 1
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
# NAND 게이트를 구현하시오. 입력 2개 가중치 2개(w1= -0.5, w2= -0.5), bias = 0.7로 선형 결합 결과가 0보다 작으면 0, 아니면 1
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
# OR 게이트를 구현하시오. 입력 2개 가중치 2개(w1= 0.5, w2= 0.5), bias = -0.2로 선형 결합 결과가 0보다 작으면 0, 아니면 1
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# sigmoid 함수를 구현하시오.
# np.exp() 함수를 이용하여 e의 지수함수를 구할 수 있음.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
t = np.array([-1.0, 1.0, 2.0])
sigmoid(t)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.show()
```
